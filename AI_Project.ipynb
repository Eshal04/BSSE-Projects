# Remove all null values in numerical columns

```python
Eshal.isna()
```

```python
for i in Eshal.columns:
    if (Eshal[i].dtype=='float64')|(Eshal[i].dtype=='int64'):
        Eshal[i]=Eshal[i].replace(np.nan,Eshal[i].mean())
```

```python
Eshal.info()
```

# Balance All samples

```python
Eshal=Eshal.dropna(subset='continent')
```

```python
Eshal=Eshal.drop('tests_units',axis=1)
```

```python
Eshal.info()
```

# Convert null values with Arid number

```python
for i in Eshal.columns:
    if(Eshal[i].dtype=='float64')|(Eshal[i].dtype=='int64'):
        Eshal[i]=Eshal[i].replace([0.0],[4655])
```
